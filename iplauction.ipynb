```python
bowler_count = df_sold_bowler.groupby('2023 Squad')['Player Name'].count().reset_index()
bowler_count = bowler_count.rename(columns={'Player Name': 'Bowler Count'})
team_spend = df_sold_bowler.groupby('2023 Squad')['Cost in Rs. (CR)'].sum().reset_index()

team_summary = pd.merge(bowler_count, team_spend, on='2023 Squad')

team_summary['Avg Cost per Bowler (CR)'] = team_summary['Cost in Rs. (CR)'] / team_summary['Bowler Count']

plt.figure(figsize=(10, 6))
sns.barplot(data=team_summary, x='2023 Squad', y='Bowler Count', palette='Blues')
plt.title('Number of Bowlers Bought by Each Team in 2023')
plt.xlabel('2023 Squad')
plt.ylabel('Number of Bowlers')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(data=team_summary, x='2023 Squad', y='Avg Cost per Bowler (CR)', palette='Greens')
plt.title('Average Cost per Bowler Bought by Each Team in 2023')
plt.xlabel('2023 Squad')
plt.ylabel('Average Cost per Bowler (in Crores)')
plt.xticks(rotation=45)
plt.show()
```
