```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("ipl_2023_dataset.csv")
```

```python
df.head()
```

```python
df_batsman=df[df['Type']=='BATSMAN']
```

```python
df_batsman.shape
```

```python
df_batsman
```

```python
df_batsman=df_batsman[df['Base Price']!='Retained']
```

```python
df_batsman
```

```python
df_batsman_unsold=df_batsman[df_batsman['2023 Squad']=='Unsold']
```

```python
df_batsman.drop(df_batsman[df_batsman['2023 Squad'] == 'Unsold'].index, inplace=True)
```

```python
df_batsman
```

```python
df_batsman['Base Price']=df_batsman['Base Price'].astype(int)
```

```python
df_batsman
```

```python
df_batsman['Base Price']=df_batsman['Base Price']//100000
```

```python
df_batsman
```

```python
df_batsman['Cost in Rs. (CR)']=df_batsman['Cost in Rs. (CR)']*100
```

```python
df_batsman
```

```python
df_batsman=df_batsman.drop(columns=['Cost in $ (K)','2022 Squad'])
```

```python
df_batsman
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(12, 6))
plt.bar(df_batsman['Player Name'], df_batsman['Cost in Rs. (CR)'], color='skyblue')
plt.xticks(rotation=45, ha='right')
plt.title('Player Costs in Crores (Rs.)')
plt.xlabel('Players')
plt.ylabel('Cost (Rs. CR)')
plt.show()
```

```python
teams=df_batsman['2023 Squad']
team_counts = pd.Series(teams).value_counts()

# Plot
plt.figure(figsize=(8, 8))
team_counts.plot.pie(autopct='%1.1f%%', colors=plt.cm.Paired.colors, startangle=140)
plt.title('Team Composition')
plt.ylabel('')  # Remove y-axis label
plt.show()
```

```python

team_costs = df_batsman.groupby('2023 Squad')['Cost in Rs. (CR)'].sum()

plt.figure(figsize=(8, 6))
team_costs.sort_values().plot(kind='barh', color='orange')
plt.title('Total Spending by Teams')
plt.xlabel('Total Cost (Rs. CR)')
plt.ylabel('Teams')
plt.show()
```
